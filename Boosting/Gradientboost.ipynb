```python
import pandas as pd 
dataset=pd.read_csv("insurance.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset=dataset.astype(int)
dataset
```

```python
indep=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes',
       'region_northwest', 'region_southeast', 'region_southwest']]
```

```python
dep=dataset[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
train_test_split(indep,dep,test_size=0.30,random_state=0)
x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=0.30,random_state=0)
```

```python
# data preprocessing
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
regr=GradientBoostingRegressor(loss='squared_error', learning_rate=0.1, min_samples_split=2)
```

```python
regr.fit(x_train,y_train)
```

```python
y_pred=regr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
import pickle
file="finalized_model_gradientboosting.sav"
pickle.dump(regr,open(file,'wb'))
```

```python
loaded_model=pickle.load(open("finalized_model_gradientboosting.sav",'rb'))
```
